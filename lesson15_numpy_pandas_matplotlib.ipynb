# NumPy + Pandas + Matplotlib 基础

深度学习三大基础工具:
- **NumPy**: 数值计算, 张量运算的基础
- **Pandas**: 数据处理与分析
- **Matplotlib**: 数据可视化

它们的关系:
```
原始数据 -> Pandas(清洗/处理) -> NumPy(数值计算) -> Matplotlib(可视化)
                                              -> PyTorch(深度学习)
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

# 一、NumPy

## 1.1 创建数组

````python
# 从列表创建
a = np.array([1, 2, 3, 4, 5])
print("一维数组:", a, "类型:", a.dtype, "形状:", a.shape)

b = np.array([[1, 2, 3], [4, 5, 6]])
print("二维数组:\n", b, "形状:", b.shape)

# 指定数据类型
c = np.array([1, 2, 3], dtype=np.float32)
print("float32:", c, c.dtype)
````

````python
# 特殊数组
print("全零:", np.zeros(5))
print("全零2D:\n", np.zeros((2, 3)))
print("全一:", np.ones((2, 3)))
print("单位阵:\n", np.eye(3))
print("等差:", np.arange(0, 10, 2))
print("等间隔:", np.linspace(0, 1, 5))
print("随机均匀:", np.random.rand(3))
print("随机正态:", np.random.randn(3))
print("随机整数:", np.random.randint(0, 10, size=5))
````

## 1.2 数组属性与形状操作

````python
a = np.random.randn(3, 4)
print(f"形状: {a.shape}, 维度: {a.ndim}, 元素数: {a.size}, 数据类型: {a.dtype}")

# reshape
b = a.reshape(2, 6)
print(f"reshape (2,6): {b.shape}")

# -1 自动推断
c = a.reshape(3, -1)  # 3行, 列自动算
print(f"reshape (3,-1): {c.shape}")

# 展平
print(f"flatten: {a.flatten().shape}, ravel: {a.ravel().shape}")

# 转置
print(f"转置: {a.T.shape}")

# 增加维度
print(f"np.newaxis: {a[np.newaxis].shape}")  # (1, 3, 4)
print(f"expand_dims: {np.expand_dims(a, axis=0).shape}")
````

## 1.3 索引与切片

````python
a = np.arange(12).reshape(3, 4)
print("数组:\n", a)
print()
print("第0行:", a[0])
print("第1列:", a[:, 1])
print("第1行第2列:", a[1, 2])
print("前两行:", a[:2])
print("步长切片:", a[::2])  # 每隔一行
print()

# 布尔索引
print("大于5:", a[a > 5])

# 花式索引
print("取第0行和第2行:", a[[0, 2]])
print("取(0,1)和(2,3):", a[[0, 2], [1, 3]])
````

## 1.4 数学运算

````python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

# 逐元素运算
print("加:", a + b)
print("乘:", a * b)      # 逐元素, 不是矩阵乘法
print("幂:", a ** 2)
print("标量运算:", a * 10 + 1)
print()

# 通用函数 (ufunc)
print("sqrt:", np.sqrt(a))
print("exp:", np.exp(a))
print("log:", np.log(a))
print("abs:", np.abs([-1, -2, 3]))
print("sin:", np.sin(np.array([0, np.pi/2, np.pi])))
````

````python
# 聚合运算
a = np.random.randn(3, 4)
print("数组:\n", a.round(2))
print()
print("总和:", a.sum())
print("按行求和:", a.sum(axis=1))    # 每行求和
print("按列求和:", a.sum(axis=0))    # 每列求和
print("均值:", a.mean())
print("标准差:", a.std())
print("最大值:", a.max())
print("最小值索引:", a.argmin())
print()
print("axis=0 沿行方向(列聚合), axis=1 沿列方向(行聚合)")
````

````python
# 矩阵运算
A = np.random.randn(3, 4)
B = np.random.randn(4, 2)

print("矩阵乘法 A@B:\n", (A @ B).round(2))
print("矩阵乘法 dot:\n", np.dot(A, B).round(2))
print()

# 内积、外积
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("内积:", np.dot(a, b))
print("外积:\n", np.outer(a, b))
````

## 1.5 广播机制

````python
# 广播: 不同形状的数组运算时自动扩展
a = np.array([[1, 2, 3],
              [4, 5, 6]])   # (2, 3)
b = np.array([10, 20, 30])  # (3,)

print("(2,3) + (3,):\n", a + b)  # b 自动扩展为2行
print()

c = np.array([[10], [20]])  # (2, 1)
print("(2,3) + (2,1):\n", a + c)  # c 自动扩展为3列
print()

# 标量广播
print("(2,3) + 标量:\n", a + 100)
````

## 1.6 NumPy 与 PyTorch 互转

````python
import torch

# NumPy -> PyTorch
np_arr = np.array([1, 2, 3])
torch_tensor = torch.from_numpy(np_arr)
print(f"NumPy -> Tensor: {np_arr} -> {torch_tensor}")

# PyTorch -> NumPy
torch_arr = torch.tensor([4.0, 5.0, 6.0)
np_back = torch_arr.numpy()
print(f"Tensor -> NumPy: {torch_arr} -> {np_back}")

# 共享内存! 修改一个另一个也变
np_arr[0] = 999
print(f"修改 numpy 后 tensor 也变了: {torch_tensor}")
print()
print("注意: 共享内存仅限 CPU Tensor, GPU Tensor 转换会复制数据")
````

# 二、Pandas

## 2.1 Series 与 DataFrame

````python
# Series: 一维数据
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'], name='score')
print("Series:")
print(s)
print(f"\n值: {s.values}, 索引: {s.index}")
````

````python
# DataFrame: 二维表格
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'age': [25, 30, 35, 28, 22],
    'score': [85, 92, 78, 95, 88],
    'city': ['Beijing', 'Shanghai', 'Guangzhou', 'Shenzhen', 'Hangzhou']
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
print(f"\n形状: {df.shape}, 列名: {df.columns.tolist()}")
````

````python
# 查看数据
print("=== 数据概览 ===")
print(f"\n前3行:\n{df.head(3)}")
print(f"\n后2行:\n{df.tail(2)}")
print(f"\n数据类型:\n{df.dtypes}")
print(f"\n统计描述:\n{df.describe()}")
print(f"\n信息:")
df.info()
````

## 2.2 选择数据

````python
# 选择列
print("单列 (Series):")
print(df['name'])
print()
print("多列 (DataFrame):")
print(df[['name', 'score']])
````

````python
# 选择行
print("按位置 iloc:")
print(df.iloc[0])       # 第0行
print()
print(df.iloc[1:3])     # 第1-2行
print()
print("按标签 loc:")
print(df.loc[0])        # 标签为0的行
print()
print(df.loc[0:2, ['name', 'score']])  # 行标签+列名
````

````python
# 条件筛选
print("年龄>25:")
print(df[df['age'] > 25])
print()
print("分数85-90之间:")
print(df[(df['score'] >= 85) & (df['score'] <= 90)])
print()
print("城市在列表中:")
print(df[df['city'].isin(['Beijing', 'Shanghai'])])
````

## 2.3 数据处理

````python
# 添加列
df['passed'] = df['score'] >= 85
df['grade'] = df['score'].apply(lambda s: 'A' if s>=90 else ('B' if s>=80 else 'C'))
print("添加列后:")
print(df)
````

````python
# 排序
print("按分数降序:")
print(df.sort_values('score', ascending=False))
````

````python
# 分组聚合
print("按城市分组统计:")
grouped = df.groupby('city').agg(
    avg_score=('score', 'mean'),
    avg_age=('age', 'mean'),
    count=('name', 'count')
)
print(grouped)
````

````python
# 缺失值处理
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})
print("含缺失值:\n", df_missing)
print()
print("检测缺失:\n", df_missing.isnull())
print()
print("每列缺失数:\n", df_missing.isnull().sum())
print()
print("删除缺失行:\n", df_missing.dropna())
print()
print("填充缺失:\n", df_missing.fillna(0))
print()
print("用均值填充:\n", df_missing.fillna(df_missing.mean()))
````

````python
# 读写文件
print("=== 文件读写 ===")
print()
print("读取 CSV:")
print("  df = pd.read_csv('data.csv')")
print()
print("写入 CSV:")
print("  df.to_csv('output.csv', index=False)")
print()
print("读取 Excel:")
print("  df = pd.read_excel('data.xlsx')")
print()
print("其他格式:")
print("  pd.read_json(), pd.read_sql(), pd.read_parquet()")
````

````python
# Pandas -> NumPy -> PyTorch
print("=== 数据转换链路 ===")
print()
print("Pandas -> NumPy:")
arr = df[['age', 'score']].values  # 只取数值列
print(f"  df.values: shape={arr.shape}, dtype={arr.dtype}")
print()
print("NumPy -> PyTorch:")
tensor = torch.from_numpy(arr.astype(np.float32))
print(f"  torch.from_numpy: shape={tensor.shape}, dtype={tensor.dtype}")
print()
print("完整链路: CSV -> DataFrame -> NumPy -> Tensor -> 模型训练")
````

# 三、Matplotlib

## 3.1 基础折线图

````python
x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize=(8, 4))
plt.plot(x, np.sin(x), label='sin(x)', linewidth=2)
plt.plot(x, np.cos(x), label='cos(x)', linewidth=2, linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric Functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
````

## 3.2 子图 (Subplots)

````python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 折线图
axes[0, 0].plot(x, np.sin(x), 'b-', linewidth=2)
axes[0, 0].set_title('Line Plot')
axes[0, 0].grid(True, alpha=0.3)

# 散点图
np.random.seed(42)
sx = np.random.randn(50)
sy = np.random.randn(50)
colors = np.random.rand(50)
axes[0, 1].scatter(sx, sy, c=colors, cmap='viridis', alpha=0.7)
axes[0, 1].set_title('Scatter Plot')

# 柱状图
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 25, 10]
axes[1, 0].bar(categories, values, color=['#2196F3', '#4CAF50', '#FF9800', '#E91E63'])
axes[1, 0].set_title('Bar Plot')

# 直方图
data = np.random.randn(1000)
axes[1, 1].hist(data, bins=30, color='steelblue', edgecolor='white', alpha=0.7)
axes[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
````

## 3.3 深度学习常用图表

````python
# 模拟训练过程
np.random.seed(42)
epochs = np.arange(1, 51)
train_loss = 2.0 * np.exp(-0.06 * epochs) + 0.1 + np.random.randn(50) * 0.02
val_loss = 2.0 * np.exp(-0.05 * epochs) + 0.15 + np.random.randn(50) * 0.03
train_acc = 1 - train_loss / 2.0
val_acc = 1 - val_loss / 2.0

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Loss 曲线
axes[0].plot(epochs, train_loss, label='Train Loss', linewidth=2)
axes[0].plot(epochs, val_loss, label='Val Loss', linewidth=2)
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].set_title('Training & Validation Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Accuracy 曲线
axes[1].plot(epochs, train_acc, label='Train Acc', linewidth=2)
axes[1].plot(epochs, val_acc, label='Val Acc', linewidth=2)
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].set_title('Training & Validation Accuracy')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

````python
# 混淆矩阵可视化
from matplotlib.colors import LinearSegmentedColormap

cm = np.array([[45, 2, 3],
               [1, 50, 1],
               [2, 3, 43]])
classes = ['Cat', 'Dog', 'Bird']

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(cm, cmap='Blues')
ax.set_xticks(range(3))
ax.set_yticks(range(3))
ax.set_xticklabels(classes)
ax.set_yticklabels(classes)
ax.set_xlabel('Predicted')
ax.set_ylabel('True')
ax.set_title('Confusion Matrix')

for i in range(3):
    for j in range(3):
        color = 'white' if cm[i, j] > 25 else 'black'
        ax.text(j, i, str(cm[i, j]), ha='center', va='center', color=color, fontsize=14)

plt.colorbar(im)
plt.tight_layout()
plt.show()
````

````python
# 特征分布可视化
np.random.seed(42)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 正态分布
axes[0].hist(np.random.randn(1000), bins=30, color='steelblue', alpha=0.7)
axes[0].set_title('Normal Distribution')
axes[0].axvline(0, color='red', linestyle='--', label='mean')
axes[0].legend()

# 均匀分布
axes[1].hist(np.random.rand(1000), bins=30, color='coral', alpha=0.7)
axes[1].set_title('Uniform Distribution')

# 指数分布
axes[2].hist(np.random.exponential(1, 1000), bins=30, color='seagreen', alpha=0.7)
axes[2].set_title('Exponential Distribution')

for ax in axes:
    ax.set_xlabel('Value')
    ax.set_ylabel('Count')

plt.tight_layout()
plt.show()
````

````python
# 热力图 (特征相关性)
np.random.seed(42)
data = np.random.randn(100, 5)
data[:, 2] = data[:, 0] * 0.8 + np.random.randn(100) * 0.3  # 让第3列和第1列相关
corr = np.corrcoef(data.T)
labels = ['feat1', 'feat2', 'feat3', 'feat4', 'feat5']

fig, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(corr, cmap='RdBu_r', vmin=-1, vmax=1)
ax.set_xticks(range(5))
ax.set_yticks(range(5))
ax.set_xticklabels(labels)
ax.set_yticklabels(labels)
ax.set_title('Feature Correlation Matrix')

for i in range(5):
    for j in range(5):
        ax.text(j, i, f'{corr[i,j]:.2f}', ha='center', va='center', fontsize=10)

plt.colorbar(im)
plt.tight_layout()
plt.show()
````

## 总结

### NumPy 速查

| 操作 | 代码 |
|------|------|
| 创建 | `np.array()`, `zeros()`, `ones()`, `arange()`, `linspace()` |
| 形状 | `reshape()`, `flatten()`, `T`, `np.newaxis` |
| 索引 | `[i]`, `[:, j]`, `[mask]`, `[[0,2]]` |
| 运算 | `+ - * / **`, `sum()`, `mean()`, `max()`, `@` |
| 随机 | `rand()`, `randn()`, `randint()` |

### Pandas 速查

| 操作 | 代码 |
|------|------|
| 创建 | `pd.DataFrame(dict)`, `pd.read_csv()` |
| 查看 | `head()`, `info()`, `describe()` |
| 选择 | `df['col']`, `df.iloc[]`, `df.loc[]`, `df[条件]` |
| 处理 | `fillna()`, `dropna()`, `groupby()`, `apply()`, `sort_values()` |
| 转换 | `.values` -> NumPy, `.to_csv()` -> 文件 |

### Matplotlib 速查

| 图表 | 代码 |
|------|------|
| 折线图 | `plt.plot(x, y)` |
| 散点图 | `plt.scatter(x, y)` |
| 柱状图 | `plt.bar(x, y)` |
| 直方图 | `plt.hist(data)` |
| 热力图 | `plt.imshow(matrix)` |
| 子图 | `plt.subplots(2, 2)` |

### 数据处理完整链路

```
CSV文件
  -> pd.read_csv()        # Pandas 读取
  -> df.dropna()/fillna()  # 清洗数据
  -> df[条件]              # 筛选数据
  -> df.values             # 转 NumPy
  -> torch.from_numpy()    # 转 PyTorch
  -> model(tensor)         # 训练模型
  -> plt.plot(losses)      # 可视化结果
```
